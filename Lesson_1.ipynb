```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 10))

x = np.linspace(0, 10, 1000)
k1 = 1
k2 = 12

plt.plot(x, np.cos(k1 * x), color='red')
plt.plot(x, np.cos(k2 * x), color='blue', alpha=0.25)


```
